## Dealing with images that need preprocessing
#### Preprocessing is Image specific. The same parameters do not work on other images. 

```python
image = cv2.imread('ss7.png')
text = tess.image_to_string(image, lang = 'eng')
image2 = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image2)
print(text)

# As seen in the output, the text isn't captured at all. 
```

#### Much better result after preprocessing

```python
image2 = cv2.imread('ss7.png')
image2 = cv2.cvtColor(image2,cv2.COLOR_BGR2GRAY)
image2 = cv2.Canny(image2,10,100)
kernel = np.ones((5,2),np.uint8)
image2 = cv2.dilate(image2, kernel, iterations = 1)
text = tess.image_to_string(image2, lang = 'eng')
print(text)
plt.imshow(image2.astype(np.uint8),cmap = 'gray')
```

### Capturing the lyrics of gray scale frames

```python
video = cv2.VideoCapture('video8.mp4')
```

```python
idx = 0

with open('lyrics.txt','w') as file:
    while video.isOpened():
        ret,frame = video.read()
        if not ret:
            break
        name = './frames/frame' + str(idx) + '.jpg'
        if(idx%250 == 0): #creating a delay
            text = tess.image_to_string(frame, lang = 'eng')
            print(text)
            #frame = preprocess(frame)
            cv2.imwrite(name,frame)
            file.write(text + str('\n'))
        prev = text.lower()
        idx = idx +1
    video.release()
    cv2.destroyAllWindows()
```
